```python
import requests
from bs4 import BeautifulSoup
import re
```

```python
url = "https://www.dotabuff.com/heroes/abaddon/counters?date=patch_7.32"
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3"}
req = requests.get(url, headers=headers)
```

```python
soup = BeautifulSoup(req.content, "html.parser")
```

```python
header = soup.find("header", string="Matchups")
header
```

```python
table = header.find_next_sibling("article").find("table")
```

```python
for row in table.find_all("tr"):
    hero_name_tag = row.find("a", class_="link-type-hero")
    if row.find("td") and hero_name_tag:
        disadvantage = row.find_all("td")[2]['data-value']
        wr = row.find_all("td")[3]['data-value']
        print(hero_name_tag.text.strip(), disadvantage, wr)
```

```python
url2 = "https://www.dotabuff.com/heroes"
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3"}

req = requests.get(url2, headers=headers)
```

```python
soup = BeautifulSoup(req.content, "html.parser")
```

```python
hero_links = soup.find_all("a", href=re.compile(r"/heroes/.*"))
```

```python
hero_names=[]
for link in hero_links:
    if link.find("div", class_="hero"):
        hero_names.append({"hero_name": link.text.strip(), "link_name": link['href'][8:]})

hero_names
```
